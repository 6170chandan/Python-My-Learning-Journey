**Programming Language**:
The language which is used to communicate with the computer is known as programming language.
ex. Java, Golang, Julia, C, C++, FORTRAN etc.
 
**Levels of Programming**

1) **Low Level Language**: The language which is understandable by the machine is low level language or machine level language.
EX. Binary Language 

2) **Assembly level language**(Bit Level Language): 
The language which is understandable by Micriprocessor or microcontroller is known as assembly level language or bit level language. 
In assembly level language there are predefined words which is used to perform the task.
Ex. Sum, Ack, Sub etc..

The assembly language level language will be understandable by the machine by using a translate called as assembler.
 

 Assembler is a software which is used to convert assembly level language to low level language.

**Python Achitecture**

The high level language ie Python code or instructions (source code) is converted and checked with correct format and syntax by compiler.

Python compiler converts the code to **byte code**and checks for the errors and sends it to the hardware(CPU).

The byte code is again passes through **Python Virtual Machine** which is a **Software** inside it we have **Interpreter** which checks the error(s) line by line.

If PVM(Python Virtual Machine) encounters any error(s) than it stops the execution of code.

Interpreter converts the byte code to machine level code.
